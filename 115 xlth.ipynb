```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
Fs = 5000
```

```python
n = np.arange(0,100)
print(n)
```

```python
t = n/Fs
```

Cau a)

F0 = 0.5kHz

```python
F1 = 500
```

```python
xn1 = np.sin(2*math.pi*F1/Fs*n)
```

```python
plt.figure(figsize=(30,10))
markerline, stemlines, baseline = plt.stem(n, xn1, '-.')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)
plt.show()
```

F2 = 2kHz

```python
F2 = 2000
```

```python
xn2 = np.sin(2*math.pi*F2/Fs*n)
plt.figure(figsize=(30,10))
markerline, stemlines, baseline = plt.stem(n, xn2, '-.')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)
plt.show()
```

F4 = 4.5kHz

```python
F4 = 4500
```

```python
xn4 = np.sin(2*math.pi*F4/Fs*n)
plt.figure(figsize=(30,10))
markerline, stemlines, baseline = plt.stem(n, xn4, '-.')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)
plt.show()
```

xn1 và xn2 là hai tín hiệu rời rạc khác nhau, và f1, f2 được sử dụng để lấy mẫu trong thực tế
xn1 và xn4 có f4-fs/2=fs/2-f1, gây ra hiện tượng trùm phổ, tạo hai tín hiệu rời rạc như nhau, tương tự như vậy với xn2 và xn3
